## 1. Can you list the movies with a runtime of 180 minutes or more?

```python
#to find all the column names of the dataser
data.columns
```

```python
#TITLE OF MOVIE HAVING RUNTIME GREATER THAN 180 MINUTES
data[data['Runtime (Minutes)']>=180]
```

## 2. Which year had the highest average voting?

```python
#In which year there was a highest Voting
data.groupby('Year')['Votes'].mean().sort_values(ascending = False)
```

```python
sns.barplot(x= 'Year',y ='Votes',data =data)
plt.title("Votes by Year")
plt.show()
```

## 3. Which year had the highest average revenue?

```python
data.groupby('Year')['Revenue (Millions)'].mean().sort_values(ascending = False)
```

```python
sns.barplot(x= 'Year',y ='Revenue (Millions)',data =data)
plt.title("Revenue in millions by Year")
plt.show()
```

## 4. What is the average rating for each director?

```python
data.groupby('Director')['Rating'].mean().sort_values(ascending = False)
```

##  5. Can you list the top 10 longest movies by title and runtime?

```python
data = pd.read_csv('IMDB-Movie-Data.csv')
top_10 = data.nlargest(10,'Runtime (Minutes)')[['Title','Runtime (Minutes)']].set_index('Title')
top_10
```

```python
sns.barplot(x = 'Runtime (Minutes)',y= top_10.index,data = top_10)
plt.show()
```

## 6. How many movies were made each year?

```python
data = pd.read_csv('IMDB-Movie-Data.csv')
#data.columns
data['Year'].value_counts()
```

```python
sns.countplot(x = 'Year',data = data)
plt.title("Number Of Movies Per Year")
plt.legend('Year')
plt.show()
```

## 7. What is the title of the most popular movie (highest revenue)?

```python
data[data['Revenue (Millions)'].max() == data['Revenue (Millions)']]
```

##  8. Can you list the top 10 highest rated movies along with their directors?

```python
top10_len = data.nlargest(10,'Rating')[['Title','Rating','Director']].set_index('Title')
top10_len
```

```python
sns.barplot(x = 'Rating',y= top10_len.index, hue = 'Director', data = top10_len)
plt.show()
```

## 9. Can you list the top 10 highest grossing movies?

```python
top10_revenue = data.nlargest(10,'Revenue (Millions)')[['Title', 'Revenue (Millions)']].set_index('Title')
top10_revenue
```

```python
sns.barplot(x = 'Revenue (Millions)',y= top10_revenue.index,data = top10_revenue)
plt.show()
```
